```python
import matplotlib.pyplot as plt
import csv
import os
import pandas as pd
import numpy as np
```

```python
path = '/home/mkubita/Pulpit/Praca Magisterska/Zbiory danych/Results.xlsx'
```

```python
frame = pd.read_excel(path, sheet_name='BoT_IoT')
frame
```

```python
fig = plt.figure()
plt.scatter(frame["Dokładność(accuracy)"], frame["Czas pracy"].where(frame["Czas pracy"] == 2.15100))
plt.scatter(frame["Dokładność(accuracy)"], frame["Czas pracy"].where(frame["Czas pracy"] == 3.42500))
plt.scatter(frame["Dokładność(accuracy)"], frame["Czas pracy"].where(frame["Czas pracy"] == 4.09800))
plt.scatter(frame["Dokładność(accuracy)"], frame["Czas pracy"].where(frame["Czas pracy"] == 0.00060))
plt.scatter(frame["Dokładność(accuracy)"], frame["Czas pracy"].where(frame["Czas pracy"] == 0.02900))
plt.scatter(frame["Dokładność(accuracy)"], frame["Czas pracy"].where(frame["Czas pracy"] == 0.14020))
plt.scatter(frame["Dokładność(accuracy)"], frame["Czas pracy"].where(frame["Czas treningu"] == 0.00650), color="magenta")
plt.scatter(frame[frame["Modele - wszystkie cechy"].str.contains('MLP')]["Dokładność(accuracy)"],  frame[frame["Modele - wszystkie cechy"].str.contains('MLP')]["Czas pracy"])

plt.legend(["AdaBoost", "GRU", "LSTM", "Drzewo Decyzyjne", "Las Losowy", "GBT", "GaussianNB", "Modele sieci neuronowych (MLP)"])
plt.xlabel("Dokładność klasyfikacji [%]")
plt.ylabel("Czas klasfyikacji [s]")
fig.set_figheight(8)
fig.set_figwidth(16)
plt.grid()
plt.show()
```
